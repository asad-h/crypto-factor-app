## Limitations
- Survivorship bias remains because the asset master is seeded from the current mapped universe.
- Binance USD-M shortability is treated as a current shortability proxy, not a full historical listing calendar.
- Public CoinGecko and DefiLlama coverage is sparse for some assets and can lag exchange prices.
- Binance open-interest history is recent-history limited, so OI signals rely mostly on DefiLlama coverage.
- Regime-specific samples are smaller than global samples, especially once label completeness and universe filters are applied.
